```python
import requests

# Example function to get the organism from UniProt based on the protein ID
def get_organism_from_uniprot(target_id):
    url = f'https://www.uniprot.org/uniprot/{target_id}.xml'
    response = requests.get(url)
    print(response.status_code)
    if response.status_code == 200:
        # Parse XML and extract organism
        from xml.etree import ElementTree as ET
        tree = ET.ElementTree(ET.fromstring(response.content))
        root = tree.getroot()
        organism = root.find(".//organism/name[@type='scientific']").text
        return organism
    else:
        return None

# Example: Find organism for missing target IDs
organism = get_organism_from_uniprot('Plasma kallikrein')
print(f'Target ID: {'Plasma kallikrein'}, Organism: {organism}')
```

```python
source_research_num = df_with_target_sources["Target Source Organism According to Curator or DataSource"].value_counts()
```

```python
df_sources_reasearch = pd.DataFrame(list(source_research_num.items()), columns=['Source', 'Research count'])
```

```python
sources_reasearch_counts_chart = dict()
for index, row in df_sources_reasearch.head(10).iterrows() :
    sources_reasearch_counts_chart[row["Source"]] = row["Research count"]
sources_reasearch_counts_chart["Other"] = df_sources_reasearch.iloc[10:]["Research count"].sum() 
```

```python
# Create a pie chart
plt.figure(figsize=(6, 6))  # Optional: set the figure size
plt.pie(sources_reasearch_counts_chart.values(), labels=sources_reasearch_counts_chart.keys(), autopct='%1.1f%%', startangle=90)

# Show the plot
plt.title("Proportions of experiments given TSO")
plt.show()
```
